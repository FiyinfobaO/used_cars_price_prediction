#####Preprocessing the Color column

```python
df.color.value_counts()
```

```python
# convert all colors to lower case
df.color = df.color.str.lower()
```

```python
df.color.value_counts()
```

```python
df.color.unique()
```

```python
# replace ambiguous color names to simpler ones
# black
df.color = df.color.replace(['not registered', 'black (black)', 'black (black )', 'black (repainted )', 'black (repainted)', 'black (partly resprayed)', 'black (firstbody)', 'metalic black'], 'black')
# blue
df.color = df.color.replace(['blue (blue)', 'blue (repainted )', 'blue (blue black)', 'blue (repainted)', 'bule', 'blue (blue )', 'blue (sky blue)', 'blue (light blue)'], 'blue')
# grey
df.color = df.color.replace(['gray', 'grey (grey)', 'grey (repainted)', 'grey (grey )'], 'grey')
# green
df.color = df.color.replace(['dark green', 'green (repainted )', 'green (green)', 'green (first body )', 'green (dark green)', 'green (repainted)', 'green (light green)'], 'green')
# beige
df.color = df.color.replace(['beidge (beige)', 'aqua green'], 'beige')
# silver
df.color = df.color.replace(['sivler', 'sliver', 'silver (repainted)', 'silver (first body )', 'silver (silver )', 'silver (silver)', 'silver (repainted )', 'silver (partly resprayed)'], 'silver')
# white
df.color = df.color.replace(['off white', 'white (white)', 'white (repainted)', 'white (white )'], 'white')
# gold
df.color = df.color.replace(['gold (gold)', 'gold (partly resprayed)', 'gold (repainted)'], 'gold')
# brown
df.color = df.color.replace(['brown (brown)', 'brown (coffee brown)', 'brown (repainted)', 'brown (light brown)'], 'brown')
# purple
df.color = df.color.replace(['purple (purple)', 'violet'], 'purple')
# yellow
df.color = df.color.replace(['yellow (yellow)'], 'yellow')
df.loc[1404, 'color'] = 'wine'
# wine
df.color = df.color.replace(['red (wine)', 'red (wine red)', 'wine colour', 'burgundy'], 'wine')
# red
df.color = df.color.replace(['wine red', 'red (tomatoe red)', 'red (red)', 'red (repainted)', 'red wine'], 'red')
```
